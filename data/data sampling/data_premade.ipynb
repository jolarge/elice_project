```python
import pandas as pd
from tqdm import tqdm
```

```python
import pymongo
from pymongo import MongoClient

def mongo_connect(host, port, db_name, db_col):
    client = MongoClient(host, port)
    db = client[db_name]
    col = db[db_col]
    return col

col = mongo_connect('mongodb://localhost', 27017, 'newsDB', 'sample')
```

```python
def mongo_data_insert_many(host, port, db_name, db_col, json_data):
    client = MongoClient(host, port)
    db = client[db_name]
    col = db[db_col]
    # 혹시 cp848 error 가 뜨시는 분은 open(json_file_name, 'r',  encoding = 'utf-8')로 바꿔주세요.
    # with open(json_file_name, 'r') as f:
        # json_data = json.load(f)
    col.insert_many(json_data)
```

```python
news = pd.DataFrame(col.find())
```

```python
news
```

```python
import re
time = news['time'].map(lambda x : x.split()[0])
title = news['text_headline'].map(lambda x : re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…\"\“》’·]', ' ', x))
keyword = ['코로나', '여행', '대선', '주식', '부동산', '백신', '자영업', 'IT', '친환경', '교육', '수도권']
labels = list(news['label'])
company = news['text_company']

print(title)
u_time = list(time.unique())
u_company = list(company.unique())

print(len(u_time), len(u_company), len(keyword))

#time / keyword / company 순
#전체count / 긍정 count/ 부정 count/ noaml count / 긍정 비율 / 부정 비율 / 노말 비율
# item = [[[[0, 0, 0, 0, .0, .0, .0]] * len(u_company)]*len(keyword)]*len(u_time)

item = [[[[0, 0, 0, 0, .0, .0, .0] for _ in range(len(u_company))] for _ in range(len(keyword))] for _ in range(len(u_time))]

print(item[0][10][0][5])

time = list(time)
title = list(title)

# for time in u_time:
    # for company in u_company:
        # for word in keyword:
            # print(time, company, word)
    
for i in tqdm(range(len(time))):
    for idx in range(len(keyword)):
        if keyword[idx] in title[i]:
            # print(keyword[idx])
            t = u_time.index(time[i])
            c = u_company.index(company[i])
            item[t][idx][c][0] += 1
        
            # 긍정,부정,노말 비율
            item[t][idx][c][4] += round(labels[i][0], 2)
            item[t][idx][c][5] += round(labels[i][1], 2)
            item[t][idx][c][6] += round(labels[i][2], 2)
            
            # 긍정, 부정, 노말 count
            item[t][idx][c][1+labels[i].index(max(labels[i]))] += 1
```

```python
result = []
print(len(u_time), len(keyword), len(u_company) )

for t in range(len(u_time)):
    for k in range(len(keyword)):
        for c in range(len(u_company)):
            positive = .0
            negative = .0
            nomal = .0
            
            if item[t][k][c][0] > 0:
                nomal = item[t][k][c][4] / item[t][k][c][0] 
                negative = item[t][k][c][5] / item[t][k][c][0] 
                positive = item[t][k][c][6] / item[t][k][c][0] 
            result.append({
                            '날짜':u_time[t], '키워드':keyword[k], '언론사':u_company[c],
                            '전체':item[t][k][c][0], '노말':item[t][k][c][1],
                          '부정':item[t][k][c][2], '긍정':item[t][k][c][3],
                          '노말비율':nomal ,'부정비율':negative,
                          '긍정비율':positive})
```
