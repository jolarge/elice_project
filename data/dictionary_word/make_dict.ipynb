```python
import pandas as pd
import numpy as np
import json
import tqdm
```

```python
df1 = pd.read_json('./sentiword.json')
```

```python
df1
```

```python
df1 = df1[['word', 'polarity']]
```

```python
df1
```

```python
df2 = pd.read_csv('./senti.csv')
```

```python
df2
```

```python
positive_flag = ['Positive', 'Anticipation', 'Joy', 'Trust']
negative_flag = ['Negative', 'Anger', 'Disgust', 'Fear', 'Sadness']
value = []
label = []

for idx, item in tqdm.tqdm(df2.iterrows()):
    word = item['Korean (ko)']
    positive_value = item[positive_flag].sum()
    negative_value = item[negative_flag].sum()
    same = df1['word'].isin([word]).sum()
    if same > 0:
        continue
    value.append(word)
    if positive_value > negative_value:
        if positive_value >= 2:
            label.append(2)
        elif positive_value >= 1:
            label.append(1)
        else:
            label.append(0)
    else:
        if negative_value >= 2:
            label.append(-2)
        elif negative_value >= 1:
            label.append(-1)
        else:
            label.append(0)
        
        
        
```

```python
df3 = pd.DataFrame({'word' : value,'polarity' : label})
```

```python
df3
```

```python
result = pd.concat([df1,df3], axis = 0)
```

```python
result.to_csv('./dictionary_result.csv', index = False)
```
